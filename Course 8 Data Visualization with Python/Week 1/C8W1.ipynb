```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
apple_SP=[34.6,65.8,76.5,88.9,78.2]
MS_SP=[67.6,74.8,70.4,77.9,83.2]
year=[2014,2015,2016,2017,2018]
```

```python
plt.plot(year,apple_SP)
plt.xticks([2013,2017,2019]) # 1st method to change x and y axes
plt.yticks([30,40,60,80])
plt.show()

plt.scatter(year,MS_SP)
plt.show()
```

```python
plt.plot(year,apple_SP)
plt.xlim((2013,2019))  # 2nd method to change x and y axes
plt.ylim((20,100))
plt.show()

plt.scatter(year,MS_SP)
plt.show()
```

```python
plt.plot(year,apple_SP)
plt.axis([2013, 2019, 20, 90]) # 3rd method to change x and y axes
plt.show()

plt.scatter(year,MS_SP)
plt.show()
```

```python
plt.plot(year,apple_SP)
plt.xticks(year,["2013-2014", "2014-2015", "2015-2016", "2016-2017", "2017-2018"])
plt.show()

plt.scatter(year,MS_SP)
plt.show()
```

```python
plt.plot(year,apple_SP)
plt.xticks(year,["a", "b", "c", "d", "e"])
plt.show()

plt.scatter(year,MS_SP)
plt.show()
```

```python
plt.plot(year,apple_SP)
plt.scatter(year,MS_SP)
plt.xlabel("year")
plt.ylabel("Stock Price")
plt.show()
```

```python
plt.plot(year,apple_SP, "k",year,MS_SP, "--g") # k represents black color
plt.xlabel("year")
plt.ylabel("Stock Price")
plt.axis([2013, 2019, 20, 100])
plt.show()
```

```python
fig_1 = plt.figure(figsize=(15,10.8))

chart1 = fig_1.add_subplot(221)
chart2 = fig_1.add_subplot(222)
chart3 = fig_1.add_subplot(223)
chart4 = fig_1.add_subplot(224)

chart1.plot(year,apple_SP)
chart2.scatter(year,MS_SP)
chart3.plot(year,apple_SP)
chart4.bar(year,MS_SP)
```
